```python
#1. Write a program to identify outliers using IQR (Interquartile Range) Method on above dataframe.

import pandas as pd
file_path=(r"heart_attack_prediction_dataset.csv")
df=pd.read_csv(file_path)

selected_columns = ['Heart Rate', 'Income', 'Exercise Hours Per Week', 'BMI', 'Triglycerides']


def identify_outliers_iqr(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers


for column in selected_columns:
    outliers = identify_outliers_iqr(column)
    
    print(f"\nOutliers for '{column}':")
    print(outliers)
```

```python
#Write a program to identify outliers using Tukey's Method on above dataframe


file_path=(r"heart_attack_prediction_dataset.csv")
df=pd.read_csv(file_path)
selected_columns = ['Heart Rate', 'Income', 'Exercise Hours Per Week', 'BMI', 'Triglycerides']


def identify_outliers_tukey(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    return outliers


for column in selected_columns:
    outliers = identify_outliers_tukey(column)
    
    print(f"\nOutliers for '{column}' using Tukey's method:")
    print(outliers)
```

```python

from scipy import stats
file_path=(r"heart_attack_prediction_dataset.csv")
df=pd.read_csv(file_path)
z_score_threshold = 3
z_score = stats.zscore(df.select_dtypes(include='number'))
outliers_mask=(abs(z_score)>z_score_threshold)
outliers_df=df[outliers_mask.any(axis=1)]
print(outliers_df)
```
