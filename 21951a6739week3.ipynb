```python
import xml.etree.ElementTree as ET
import pandas as pd

def parse_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    data = []

    for student in root.findall('.//student'):
        row = {}
        for child in student:
            row[child.tag] = child.text

        data.append(row)

    return data

def xml_to_dataframe(xml_file):
    data = parse_xml(xml_file)
    df = pd.DataFrame(data)
    return df

xml_file_path = "C:\\Users\\sai raj\\OneDrive\\Desktop\\Week3\\1xml11.xml"

df = xml_to_dataframe(xml_file_path)
print(df)
```

```python
#Write a Python program to read the json file into a dataframe in python

import pandas as pd

def read_json_to_dataframe(file_path):
    try:
        df = pd.read_json(file_path)

        print("DataFrame from JSON:")
        print(df)

    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except pd.errors.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")

file_path = "C:\\Users\\sai raj\\OneDrive\\Desktop\\Week3\\1json11.json"
read_json_to_dataframe(file_path)
```

```python
#Write a Python program that reads data from a json file that may use either a comma or a semicolon as the delimiter. Parse the data and print it

import json

def parse_json_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()

            content = content.replace(';', ',')

            data = json.loads(content)

            print(data)

    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")

file_path = "C:\\Users\\sai raj\\OneDrive\\Desktop\\Week3\\1json11.json"
parse_json_file(file_path)
```

```python
#Write a Python program to convert Python dictionary object (sort by key) to JSON data

import json

def convert_dict_to_json(input_dict):
    try:
        json_data = json.dumps(input_dict, sort_keys=True, indent=2)

        print("JSON data:")
        print(json_data)

    except Exception as e:
        print(f"Error converting dictionary to JSON: {e}")

example_dict = {
    'c': 3,
    'a': 1,
    'b': 2
}

convert_dict_to_json(example_dict)
```

```python
#Write a Python program that reads data from a html file that may use either a comma or a semicolon as the delimiter. Parse the data and print it.

from bs4 import BeautifulSoup
import pandas as pd

def parse_html_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()

            soup = BeautifulSoup(content, 'html.parser')

            data_element = soup.find('div', {'id': 'data'})

            if data_element:

                rows = data_element.get_text().strip().split('\n')

                data = [row.split(';') for row in rows]

                df = pd.DataFrame(data, columns=data[0])

                print("DataFrame from HTML:")
                print(df)

    except FileNotFoundError:
        print(f"File not found: {file_path}")

file_path = "C:\\Users\\sai raj\\OneDrive\\Desktop\\Week3\\1htmlexample11.html"
parse_html_file(file_path)
```

```python
#Write a Python program to write a dataframe into the html file using a python library in python

import pandas as pd

def write_dataframe_to_html(dataframe, file_path):
    try:
        dataframe.to_html(file_path, index=False)

        print(f"DataFrame successfully written to {file_path}")

    except Exception as e:
        print(f"Error writing DataFrame to HTML: {e}")

example_data = {
    'Name': ['John', 'Alice', 'Bob'],
    'Age': [25, 30, 28],
    'Country': ['USA', 'Canada', 'UK']
}

df = pd.DataFrame(example_data)

output_file_path = "output.html"

write_dataframe_to_html(df, output_file_path)
```
