```python
"""4.2. SORTING"""
import pandas as pd

# Example dataset as a dictionary
data = {
    'make': ['Toyota', 'Ford', 'Chevrolet', 'Nissan', 'Mercedes-Benz', 'Audi', 'BMW', 'Honda', 'Hyundai', 'Volkswagen'],
    'model': ['Camry', 'F-150', 'Silverado', 'Altima', 'C-Class', 'A4', '3 Series', 'Civic', 'Elantra', 'Jetta'],
    'year': [2022, 2022, 2022, 2022, 2022, 2022, 2022, 2022, 2022, 2022]
}

# Create DataFrame from dictionary
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)
print()

# Sort the DataFrame based on the 'make' column with a maximum length of 50 characters
df_sorted = df.sort_values(by='model', key=lambda x: x.str[:50])

# Display the sorted DataFrame
print("DataFrame sorted by 'make' (up to 50 characters):")
print(df_sorted)

```

```python
"""4.2.4. Write a program to obtain a new dataframe to the existing dataframe, To apply function to every cell of
dataframe."""

import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(df)
print()

# Define a function to apply to each cell
def square_cell(value):
    return value ** 2

# Apply the function to every cell in the DataFrame
df_squared = df.applymap(square_cell)

# Display DataFrame with squared values
print("DataFrame with squared values:")
print(df_squared)
```

```python
""" 4.4. Write a program for performing aggregation by grouping by a column ‘Frequency of Purchases&#39; and
calculate sum and mean of ‘Review Rating&#39; column from the above dataframe."""

import pandas as pd

# Sample DataFrame
data = {
    'Frequency of Purchases': ['High', 'Low', 'Medium', 'High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High'],
    'Review Rating': [4, 3, 5, 4, 2, 3, 5, 4, 2, 5]
}

df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(df)
print()

# Group by 'Frequency of Purchases' and calculate sum and mean of 'Review Rating'
aggregated_df = df.groupby('Frequency of Purchases')['Review Rating'].agg(['sum', 'mean'])

# Display the aggregated DataFrame
print("Aggregated DataFrame:")
print(aggregated_df)
```

```python
"""4.4. 1. Write a program for performing aggregation by grouping by a column ‘Frequency of Purchases&#39; and
calculate count, min and max of ‘Review Rating&#39; column from the above dataframe."""

import pandas as pd

# Sample DataFrame
data = {
    'Frequency of Purchases': ['High', 'Low', 'Medium', 'High', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High'],
    'Review Rating': [4, 3, 5, 4, 2, 3, 5, 4, 2, 5]
}

df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(df)
print()

# Group by 'Frequency of Purchases' and calculate count, min, and max of 'Review Rating'
aggregated_df = df.groupby('Frequency of Purchases')['Review Rating'].agg(['count', 'min', 'max'])

# Display the aggregated DataFrame
print("Aggregated DataFrame:")
print(aggregated_df)
```

```python
"""4.4.2. Write a program for performing aggregation by grouping by a column ‘Subscription Status’ and
calculate sum and mean of ‘Purchase Amount (USD)’ column from the above dataframe."""

# Sample DataFrame
data = {
    'Subscription Status': ['Active', 'Inactive', 'Active', 'Inactive', 'Active', 'Inactive'],
    'Purchase Amount (USD)': [50.0, 30.0, 40.0, 25.0, 60.0, 35.0]
}

df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:")
print(df)
print()

# Group by 'Subscription Status' and calculate sum and mean of 'Purchase Amount (USD)'
aggregated_df = df.groupby('Subscription Status')['Purchase Amount (USD)'].agg(['sum', 'mean'])

# Display the aggregated DataFrame
print("Aggregated DataFrame:")
print(aggregated_df)
```
