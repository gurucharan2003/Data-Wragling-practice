```python
import numpy as np
import scipy
from scipy.integrate import quad
def my_function(x):
    return x**2
a=0
b=2
result,error=quad(my_function,a,b)
print("Numerical result: ",result)
print("estimated error: ",error)
```

```python
import numpy as np
import scipy
from scipy.linalg import eig
matrix_size=2
B=np.random.rand(matrix_size, matrix_size)
eigenvalues,eigenvectors=eig(B)
print("Matrix B:")
print(B)
print("\nEigen Values:")
print(eigenvalues)
print("eigen Vectors:")
print(eigenvectors)
```

```python
import numpy as np
import scipy
from scipy.linalg import inv
matrix_size=2
c=np.random.rand(matrix_size,matrix_size)
c_inverse=inv(c)
print("Matrix C:")
print(c)
print("\n Inverse Matrix C:")
print(c_inverse)
```
