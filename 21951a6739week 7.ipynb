```python
'''7.1Write a program to create a dataframe ‘data’ containing the data from the following dataset
“https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv”.'''
import pandas as pd

# URL of the dataset
df = "Cars93_miss.csv"

# Read the dataset into a DataFrame
data = pd.read_csv(df)

# Display the first few rows of the DataFrame
print(data.head())
```

```python
'''7.1Write a program and consider given a dataframe ‘data’ with missing values, use the .dropna() method to
remove all rows containing missing values. Display the resulting dataframe. Also verify the same by printing
the summary reports. Also print the rest rows and columns summary.'''
import pandas as pd

# Assuming 'data' is the DataFrame with missing values

# Display summary report before removing missing values
print("Summary report before removing missing values:")
print(data.info())
print("\n")

# Display DataFrame before removing missing values
print("DataFrame before removing missing values:")
print(data.head())
print("\n")

# Remove rows with missing values using dropna()
data_cleaned = data.dropna()

# Display resulting DataFrame after removing missing values
print("DataFrame after removing missing values:")
print(data_cleaned.head())
print("\n")

# Display summary report after removing missing values
print("Summary report after removing missing values:")
print(data_cleaned.info())
```

```python
'''1.Write a program and consider the above dataframe with missing values, calculate the percentage of
missing values in each column and create a summary table showing the columns with the highest
percentage of missing values.'''
import pandas as pd

# Calculate percentage of missing values in each column
missing_percentages = (data.isnull().sum() / len(data)) * 100

# Create a summary table
summary_table = pd.DataFrame({
    'Column': missing_percentages.index,
    'Percentage of Missing Values': missing_percentages.values
})

# Sort summary table by percentage of missing values in descending order
summary_table = summary_table.sort_values(by='Percentage of Missing Values', ascending=False)

# Display the summary table
print(summary_table)
```

```python
'''2. Write a program and consider the above dataframe, calculate and display the number of missing values
in each column individually. Sort the columns by the number of missing values in descending order.
'''
import pandas as pd

# Assuming 'data' is the DataFrame with missing values

# Calculate number of missing values in each column
missing_count = data.isnull().sum()

# Create a DataFrame for missing value counts
missing_count_df = pd.DataFrame({
    'Column': missing_count.index,
    'Missing Values Count': missing_count.values
})

# Sort columns by number of missing values in descending order
missing_count_df = missing_count_df.sort_values(by='Missing Values Count', ascending=False)

# Display the number of missing values in each column
print("Number of missing values in each column (sorted by count in descending order):")
print(missing_count_df)
```

```python
'''3. Write a program and consider the above dataframe, calculate and display the number of missing values
in each row individually. Sort the rows by the number of missing values in ascending order.
'''
import pandas as pd

missing_count_per_row = data.isnull().sum(axis=1)


missing_count_per_row_df = pd.DataFrame({
    'Row': data.index,
    'Missing Values Count': missing_count_per_row.values
})

missing_count_per_row_df = missing_count_per_row_df.sort_values(by='Missing Values Count')


print("Number of missing values in each row (sorted by count in ascending order):")
print(missing_count_per_row_df)
```

```python
'''5.Write a program and consider above dataframe, remove rows that have more than a specified number
of missing values. Display the resulting dataframe.'''
import pandas as pd

# Assuming 'data' is the DataFrame with missing values

threshold_missing_values = 2

missing_values_per_row = data.isnull().sum(axis=1)

data_filtered = data[missing_values_per_row <= threshold_missing_values]

print("Resulting DataFrame after removing rows with more than {} missing values:".format(threshold_missing_values))
print(data_filtered)

```

```python
'''7.2 Handling missing values - Imputation
Write a program to impute the missing values in the numerical column of the given dataframe ‘data’ with
the mean of the non-missing values.'''
import pandas as pd

numerical_columns_with_missing = data.select_dtypes(include='number').columns[data.select_dtypes(include='number').isnull().any()]

for column in numerical_columns_with_missing:
    mean_value = data[column].mean()
    data[column].fillna(mean_value, inplace=True)

print("DataFrame after imputing missing values with the mean:")
print(data)

```

```python
'''7.3 Handling missing values - Interpolation
Write a program and consider the given dataframe with missing values in a numerical column, use linear
interpolation to impute the missing values in that column. '''
import pandas as pd

# Assuming 'data' is the DataFrame with missing values in a numerical column

# Replace 'YourNumericalColumn' with the column name having missing values
numerical_column = 'Min.Price'

data[numerical_column] = data[numerical_column].interpolate(method='linear')

print(data)
```

```python
'''1. Write a program to define a custom interpolation function that takes into account domain-specific
knowledge or specific data characteristics to impute missing values in a dataframe.'''
import pandas as pd
def custom_interpolation(column):
    return column.interpolate(method='linear')

# Apply the custom interpolation function to the DataFrame
data_imputed = data.apply(custom_interpolation, axis=0)

print(data_imputed)
```

```python
'''2. Write a program and consider given dataframe with time-ordered data and a numerical column
containing missing values, impute the missing values in that column. Ensure that missing values are filled
with the previously available value in the column. Perform Forward Fill using forward fill (.ffill()) method'''
import pandas as pd

# Assuming 'data' is the DataFrame with time-ordered data and missing values in a numerical column

# Replace 'YourNumericalColumn' with the column name having missing values
numerical_column = 'Min.Price'

# Impute missing values using forward fill (ffill)
data[numerical_column] = data[numerical_column].ffill()

print(data)
```
